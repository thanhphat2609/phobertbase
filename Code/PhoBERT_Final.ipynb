## Visualization

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm

def visualize_subdirectories(folder_path):
    subdirectories = []
    file_counts = []

    for root, dirs, files in os.walk(folder_path):
        if root != folder_path:
            subdirectory_name = os.path.basename(root)
            subdirectories.append(subdirectory_name)
            file_counts.append(len(files))

    subdirectories, file_counts = zip(*sorted(zip(subdirectories, file_counts), key=lambda x: x[0]))

    min_value = min(file_counts)
    max_value = max(file_counts)
    range_value = max_value - min_value
    norm = plt.Normalize(min_value - range_value * 0.2, max_value)
    cmap = plt.get_cmap('Greys')
    colors = cmap(norm(file_counts))

    plt.figure(figsize=(12, 6))
    bars = plt.barh(subdirectories, file_counts, color=colors)
    plt.xlabel('Số lượng mẩu tin')
    plt.ylabel('Thể loại')
    plt.title('Số lượng mẩu tin của mỗi thể loại trong tập' + ' ' + str(folder_path))
    plt.tight_layout()

    for bar, count in zip(bars, file_counts):
        plt.text(bar.get_width() + 50, bar.get_y() + bar.get_height() / 2, count, ha='left', va='center')

    plt.show()
```

Đoạn code trên có chức năng để trực quan hóa số lượng tập tin trong các thư mục con của một thư mục cụ thể. Dưới đây là mô tả từng phần của code:

1. Import thư viện: Đầu tiên, code import hai thư viện của Matplotlib, đó là **`matplotlib.pyplot`** và **`matplotlib.cm`**. Thư viện **`pyplot`** được sử dụng để tạo biểu đồ, và **`cm`** để xử lý màu sắc.

2. Khai báo hàm **`visualize_subdirectories`**: Hàm này nhận đầu vào là **`folder_path`** - đường dẫn tới thư mục chứa các thư mục con cần phân tích.

Các biến và giá trị này được sử dụng trong quá trình huấn luyện và đánh giá mô hình, định nghĩa các thiết lập và cấu hình quan trọng cho quá trình huấn luyện.

## Function definition

```python
def read_txt(path):
  with open(path, 'r', encoding='utf-16') as f:
    data = f.read()
  return data
```

Hàm **`read_txt(path)`** là một hàm Python nhận đầu vào là một đường dẫn tệp tin (path). Mục đích của hàm này là đọc nội dung của tệp tin văn bản (text file) có đường dẫn đã cho và trả về dữ liệu đọc được.

Cách thức hoạt động của hàm như sau:
1. Mở tệp tin với đường dẫn đã cho bằng cách sử dụng **`open(path, 'r', encoding='utf-16')`**. Trong đó:
   - **`'r'`** đại diện cho chế độ đọc (read mode), cho phép đọc dữ liệu từ tệp tin.
   - **`encoding='utf-16'`** chỉ định kiểu mã hóa (encoding) của tệp tin là UTF-16. Điều này đảm bảo rằng tệp tin với mã hóa UTF-16 có thể được đọc và xử lý chính xác.

2. Sử dụng câu lệnh **`with`** để đảm bảo rằng tệp tin được đóng tự động sau khi hoàn thành việc đọc. Khi câu lệnh **`with`** được sử dụng, không cần phải gọi phương thức **`close()`** để đóng tệp tin.

3. Đọc nội dung của tệp tin đã mở bằng cách sử dụng phương thức **`read()`** và gán kết quả vào biến **`data`**.

4. Trả về giá trị của biến **`data`**, chứa nội dung của tệp tin đã đọc.

Điều này cho phép người dùng gọi hàm **`read_txt(path)`** để đọc nội dung của một tệp tin văn bản và sử dụng dữ liệu đó trong các phần của mã Python khác.

```python
def make_data(path):
  texts = []
  labels = []
  for file_path in tqdm(glob2.glob(path)):
    try:
      content = read_txt(file_path)
      label = file_path.split('/')[1]
      texts.append(content)
      labels.append(label)
    except:
      next
  return texts, labels
```

Đoạn code trên là một hàm Python có tên **`make_data(path)`**, nhận đầu vào là một đường dẫn (path). Mục đích của hàm này là xử lý một tập hợp các tệp tin văn bản trong một thư mục và trả về danh sách các nội dung và nhãn tương ứng của chúng.

Cách thức hoạt động của hàm như sau:
1. Khởi tạo hai danh sách rỗng **`texts`** và **`labels`**, sẽ được sử dụng để lưu trữ nội dung và nhãn của các tệp tin.

2. Sử dụng vòng lặp **`for`** để lặp qua mỗi đường dẫn tệp tin trong thư mục (path) sử dụng **`glob2.glob(path)`**. Hàm **`glob2.glob()`** trả về một danh sách các đường dẫn tệp tin phù hợp với mẫu tìm kiếm (path) đã cho.

3. Trong vòng lặp, hàm sẽ thử đọc nội dung của mỗi tệp tin sử dụng hàm **`read_txt(file_path)`** và lưu trữ kết quả vào biến **`content`**. Ngoài ra, nhãn của tệp tin được trích xuất từ đường dẫn bằng cách sử dụng **`file_path.split('/')[1]`** và gán vào biến **`label`**.

4. Sau đó, nội dung và nhãn tương ứng được thêm vào danh sách **`texts`** và **`labels`** bằng cách sử dụng phương thức **`append()`**.

5. Nếu có bất kỳ lỗi nào xảy ra trong quá trình đọc tệp tin, một ngoại lệ sẽ được ném và quá trình xử lý sẽ tiếp tục với tệp tin tiếp theo bằng cách sử dụng **`except: next`**.

6. Khi vòng lặp kết thúc, danh sách **`texts`** và **`labels`** sẽ được trả về bởi hàm như kết quả của quá trình xử lý.

Điều này cho phép người dùng gọi hàm **`make_data(path)`** để xử lý tập hợp các tệp tin văn bản trong một thư mục và nhận lại danh sách nội dung và nhãn của chúng để sử dụng trong việc huấn luyện hoặc phân loại dữ liệu.

```python
def _save_pkl(path, obj):
  with open(path, 'wb') as f:
    pickle.dump(obj, f)
```

Đoạn mã trên là một hàm **`_save_pkl(path, obj)`** trong Python. Nó nhận đầu vào là một đường dẫn (path) và một đối tượng (obj), và có chức năng lưu trữ (serialize) đối tượng đó vào một tệp tin sử dụng module pickle.

Cách thức hoạt động của hàm **`_save_pkl()`** như sau:
1. Hàm sử dụng câu lệnh **`with`** để mở tệp tin với đường dẫn đã cho bằng cách sử dụng **`open(path, 'wb')`**. Trong đó:
   - **`'wb'`** đại diện cho chế độ ghi nhị phân (write binary mode), cho phép ghi dữ liệu thành chuỗi byte.
   - **`open()`** trả về một đối tượng tệp tin có thể sử dụng để ghi dữ liệu vào tệp tin.

2. Sử dụng phương thức **`pickle.dump(obj, f)`** để lưu trữ đối tượng **`obj`** vào tệp tin đã mở. Phương thức này sẽ chuyển đổi đối tượng thành chuỗi byte và ghi vào tệp tin.

3. Khi xong, câu lệnh **`with`** sẽ tự động đóng tệp tin.

Hàm **`_save_pkl()`** cho phép người dùng gọi nó để lưu trữ (serialize) một đối tượng vào một tệp tin, sử dụng module pickle. Điều này hữu ích khi bạn muốn lưu trữ và gửi đối tượng Python qua mạng, hoặc khi bạn muốn lưu trữ dữ liệu cho việc sử dụng sau này.

```python
def _load_pkl(path):
  with open(path, 'rb') as f:
    obj = pickle.load(f)
  return obj
```

Đoạn mã trên là một hàm **`_load_pkl(path)`** trong Python. Nó nhận đầu vào là một đường dẫn (path) tới tệp tin đã được lưu trữ trước đó bằng module pickle, và có chức năng khôi phục (deserialize) đối tượng từ tệp tin đó.

Cách thức hoạt động của hàm **`_load_pkl()`** như sau:
1. Hàm sử dụng câu lệnh **`with`** để mở tệp tin với đường dẫn đã cho bằng cách sử dụng **`open(path, 'rb')`**. Trong đó:
   - **`'rb'`** đại diện cho chế độ đọc nhị phân (read binary mode), cho phép đọc dữ liệu từ tệp tin dưới dạng chuỗi byte.
   - **`open()`** trả về một đối tượng tệp tin có thể sử dụng để đọc dữ liệu từ tệp tin.

2. Sử dụng phương thức **`pickle.load(f)`** để khôi phục đối tượng từ tệp tin đã mở. Phương thức này sẽ đọc chuỗi byte từ tệp tin và chuyển đổi nó thành đối tượng Python ban đầu.

3. Gán đối tượng đã khôi phục vào biến **`obj`**.

4. Khi xong, câu lệnh **`with`** sẽ tự động đóng tệp tin.

5. Hàm **`_load_pkl()`** trả về đối tượng đã khôi phục từ tệp tin.

Hàm **`_load_pkl()`** cho phép người dùng gọi nó để khôi phục (deserialize) đối tượng từ một tệp tin đã lưu trữ trước đó bằng module pickle. Điều này hữu ích khi bạn muốn khôi phục dữ liệu hoặc đối tượng đã được lưu trữ và sử dụng chúng trong mã Python của mình.

```python
max_seq_len = 256

def convert_lines(lines, vocab, bpe):
  outputs = np.zeros((len(lines), max_seq_len), dtype=np.int32)
  cls_id = 0
  eos_id = 2
  pad_id = 1

  for i, row in tqdm(enumerate(lines), total=len(lines)):
    subwords = bpe.encode('<s> ' + row + ' </s>')
    input_ids = vocab.encode_line(subwords, append_eos=False, add_if_not_exist=False).long().tolist()
    if len(input_ids) > max_seq_len:
      input_ids = input_ids[:max_seq_len]
      input_ids[-1] = eos_id
    else:
      input_ids += [pad_id,]*(max_seq_len - len(input_ids))
    
    outputs[i, :] = np.array(input_ids)
  return outputs
```

Đoạn mã trên định nghĩa một hàm **`convert_lines(lines, vocab, bpe)`** trong Python. Hàm này có chức năng chuyển đổi một danh sách các văn bản đầu vào thành ma trận đại diện (matrix representation) theo quy tắc nhất định.

Cách thức hoạt động của hàm **`convert_lines()`** như sau:

1. Đầu vào:
   - **`lines`**: là một danh sách các văn bản đầu vào.
   - **`vocab`**: là từ điển được sử dụng để mã hóa các subword.
   - **`bpe`**: là một biến không được mô tả trong đoạn mã, tuy nhiên được giả định là một phương thức mã hóa subword.

2. Khởi tạo ma trận **`outputs`** có kích thước **`(số dòng của lines, max_sequence_length, kiểu dữ liệu)`** bằng cách sử dụng hàm **`np.zeros()`** từ thư viện NumPy. Ma trận này sẽ được sử dụng để lưu trữ đầu ra của quá trình chuyển đổi.

3. Định nghĩa các biến **`cls_id`**, **`eos_id`**, và **`pad_id`** để lưu trữ chỉ mục của các subword đặc biệt, bao gồm subword đầu câu (cls), subword cuối câu (eos), và subword padding.

4. Sử dụng vòng lặp **`for`** để lặp qua từng văn bản trong danh sách **`lines`**. Trong vòng lặp, các bước sau được thực hiện:

   a. Mã hóa subwords trong văn bản theo phương pháp mã hóa byte pair encoding (bpe) bằng cách sử dụng biến **`bpe.encode()`**. Trước khi mã hóa, thêm subword đầu câu (**`<s>`**) và subword cuối câu (**`</s>`**) vào văn bản.

   b. Sử dụng từ điển **`vocab`** để mã hóa dãy subwords thành chỉ mục (**`input_ids`**) sử dụng phương thức **`vocab.encode_line()`**. Đối số **`append_eos=False`** được sử dụng để không thêm subword cuối câu vào cuối dãy chỉ mục. Đối số **`add_if_not_exist=False`** được sử dụng để không thêm subword mới vào từ điển nếu chưa tồn tại.

   c. Kiểm tra độ dài của **`input_ids`** nếu vượt quá **`max_sequence_length`**. Nếu vượt quá, cắt đi các chỉ mục thừa và gán chỉ mục cuối cùng thành chỉ mục subword cuối câu (eos_id).

   d. Nếu độ dài của **`input_ids`** không vượt quá **`max_sequence_length`**, thêm các chỉ mục padding (pad_id) vào cuối dãy chỉ mục để đạt được độ dài mong muốn **`max_sequence_length`**.

   e. Gán dãy chỉ mục đã xử lý vào hàng thứ **`idx`** trong ma trận **`outputs`**.

5. Trả về ma trận **`outputs`** là kết quả cuối cùng của quá trình chuyển đổi.

Hàm **`convert_lines()`** cho phép người dùng gọi nó để chuyển đổi danh sách các văn bản thành ma trận đại diện sử dụng quy tắc nhất định. Kết quả ma trận đại diện có thể được sử dụng để đưa vào mô hình học máy để thực hiện các tác vụ như xử lý ngôn ngữ tự nhiên, dự đoán văn bản, và nhiều tác vụ khác.

```python
def seed_everything(SEED):
  np.random.seed(SEED)
  torch.manual_seed(SEED)
  torch.cuda.manual_seed(SEED)
  torch.backends.cudnn.deterministic = True
```

Hàm **`seed_everything(SEED)`** được định nghĩa để thiết lập seed (giá trị khởi tạo) cho các phần tử ngẫu nhiên trong quá trình huấn luyện mô hình. Điều này đảm bảo rằng các phần tử ngẫu nhiên sẽ có cùng giá trị trong mỗi lần chạy, giúp kiểm soát sự tái tái hiện của quá trình huấn luyện và đảm bảo kết quả nhất quán.

Cụ thể, hàm thực hiện các công việc sau:
- **`np.random.seed(SEED)`**: Thiết lập seed cho generator ngẫu nhiên trong numpy.
- **`torch.manual_seed(SEED)`**: Thiết lập seed cho generator ngẫu nhiên trong PyTorch cho các thiết bị CPU.
- **`torch.cuda.manual_seed(SEED)`**: Thiết lập seed cho generator ngẫu nhiên trong PyTorch cho các thiết bị GPU.
- **`torch.backends.cudnn.deterministic = True`**: Đặt chế độ xác định cho các thuật toán CUDNN trong PyTorch, giúp đảm bảo sự tái tái hiện của kết quả trên các thiết bị GPU.

Bằng cách sử dụng hàm **`seed_everything(SEED)`**, có thể đảm bảo rằng quá trình huấn luyện mô hình sẽ có kết quả nhất quán và tái tái hiện được.

```python
def evaluate(logits, targets):
  logits = logits.detach().cpu().numpy()   

  y_pred = np.argmax(logits, axis = 1)
  targets = targets.detach().cpu().numpy()
  
  f1 = f1_score(targets, y_pred, average='weighted')
  acc = accuracy_score(targets, y_pred)
  return acc, f1
```

Hàm **`evaluate`** được định nghĩa để đánh giá hiệu suất của mô hình dự đoán. Dưới đây là phân tích chi tiết của hàm này:

1. **`logits = logits.detach().cpu().numpy()`**: Biến đầu vào **`logits`** là đầu ra dự đoán của mô hình. Để thực hiện tính toán và so sánh, chúng ta cần chuyển đổi **`logits`** sang kiểu dữ liệu NumPy và di chuyển dữ liệu từ GPU (nếu có) sang CPU.

2. **`y_pred = np.argmax(logits, axis=1)`**: Chúng ta sử dụng hàm **`np.argmax`** để chọn ra lớp có xác suất dự đoán cao nhất từ **`logits`**. Điều này giúp chúng ta dự đoán nhãn của mỗi mẫu dữ liệu.

3. **`targets = targets.detach().cpu().numpy()`**: Biến **`targets`** là nhãn thực tế của các mẫu dữ liệu. Chúng ta chuyển đổi nó sang kiểu dữ liệu NumPy và di chuyển dữ liệu từ GPU (nếu có) sang CPU.

4. **`f1 = f1_score(targets, y_pred, average='weighted')`**: Chúng ta sử dụng hàm **`f1_score`** từ sklearn để tính toán F1-score dựa trên **`targets`** (nhãn thực tế) và **`y_pred`** (nhãn dự đoán). Tham số **`average='weighted'`** được sử dụng để tính toán F1-score trung bình có trọng số dựa trên số lượng mẫu từng lớp.

5. **`acc = accuracy_score(targets, y_pred)`**: Chúng ta sử dụng hàm **`accuracy_score`** từ sklearn để tính toán độ chính xác (accuracy) dựa trên **`targets`** và **`y_pred`**. Hàm này tính toán tỉ lệ các mẫu được dự đoán đúng trên tổng số mẫu.

6. Hàm **`evaluate`** trả về độ chính xác (**`acc`**) và F1-score (**`f1`**) như kết quả đánh giá hiệu suất của mô hình.

```python
def validate(valid_loader, model, device):
  model.eval()
  accs = []
  f1s = []
  losses = []

  with torch.no_grad():
    for x_batch, y_batch in valid_loader:
      x_batch = x_batch.to(device)
      y_batch = y_batch.to(device)

      outputs = model.predict('new_task', x_batch)

      loss = criteria(outputs, y_batch)
      logits = torch.exp(outputs)

      acc, f1 = evaluate(logits, y_batch)
      accs.append(acc)
      
      f1s.append(f1)
      losses.append(loss.detach().cpu().numpy())
  
  mean_acc = np.mean(accs)
  mean_f1 = np.mean(f1s)
  mean_loss = np.mean(losses)
  return mean_acc, mean_f1, mean_loss
```

Hàm **`validate`** được sử dụng để đánh giá hiệu suất của mô hình trên tập dữ liệu kiểm tra. Dưới đây là phân tích chi tiết của hàm này:

1. **`model.eval()`**: Chuyển mô hình sang chế độ đánh giá, điều này đảm bảo rằng các tầng trong mô hình không được huấn luyện và sẽ sử dụng các tham số đã học để đưa ra dự đoán.

2. **`accs = [], f1s = [], losses = []`**: Khởi tạo các danh sách để lưu trữ độ chính xác (**`accs`**), F1-score (**`f1s`**) và các giá trị mất mát (**`losses`**) của từng lô dữ liệu.

3. **`with torch.no_grad():`**: Sử dụng **`torch.no_grad()`** để tắt tính toán gradient trong quá trình đánh giá mô hình, giúp tiết kiệm bộ nhớ và tăng tốc độ tính toán.

4. **`for x_batch, y_batch in valid_loader`**: Lặp qua từng lô dữ liệu trong **`valid_loader`**, trong đó **`x_batch`** là dữ liệu đầu vào và **`y_batch`** là nhãn tương ứng.

5. **`x_batch = x_batch.to(device)`**: Chuyển đổi **`x_batch`** sang thiết bị (**`device`**) được sử dụng để huấn luyện mô hình (GPU hoặc CPU).

6. **`y_batch = y_batch.to(device)`**: Chuyển đổi **`y_batch`** sang thiết bị (**`device`**) được sử dụng để huấn luyện mô hình (GPU hoặc CPU).

7. **`outputs = model.predict('new_task', x_batch)`**: Sử dụng mô hình để dự đoán nhãn (**`outputs`**) từ dữ liệu đầu vào (**`x_batch`**) cho tác vụ **`new_task`**.

8. **`loss = criteria(outputs, y_batch)`**: Tính toán giá trị mất mát (loss) giữa **`outputs`** và nhãn thực tế (**`y_batch`**) bằng cách sử dụng hàm **`criteria`** (NLLLoss).

9. **`logits = torch.exp(outputs)`**: Áp dụng hàm exponential cho **`outputs`** để tính toán xác suất dự đoán.

10. **`acc, f1 = evaluate(logits, y_batch)`**: Đánh giá độ chính xác (**`acc`**) và F1-score (**`f1`**) bằng cách sử dụng hàm **`evaluate`** với **`logits`** và **`y_batch`**.

11. **`accs.append(acc), f1s.append(f1), losses.append(loss.detach().cpu().numpy())`**: Thêm các giá trị độ chính xác, F1-score và mất mát của lô hiện tại vào danh sách tương ứng.

12. **`mean_acc = np.mean(accs), mean_f1= np.mean(f1s), mean_loss = np.mean(losses)`**: Tính giá trị trung bình của độ chính xác (**`mean_acc`**), F1-score (**`mean_f1`**) và mất mát (**`mean_loss`**) từ các danh sách đã tính toán.

13. Hàm **`validate`** trả về giá trị trung bình của độ chính xác, F1-score và mất mát trên tập dữ liệu kiểm tra.

```python
def train_on_epoch(train_loader, model, optimizer, scheduler, epoch, num_epochs, criteria, device, log_aggr = 100):
  model.train()
  sum_epoch_loss = 0
  sum_acc = 0
  sum_f1 = 0
  start = time.time()
  train_size = len(train_loader)

  for i, (x_batch, y_batch) in enumerate(train_loader):
    x_batch = x_batch.to(device)
    y_batch = y_batch.to(device)

    y_pred = model.predict('new_task', x_batch)
    logits = torch.exp(y_pred)
    acc, f1 = evaluate(logits, y_batch)
    loss = criteria(y_pred, y_batch)

    loss.backward()
    if i % ACCUMULATION_STEPS == 0 or i == train_size- 1:
      optimizer.step()
      optimizer.zero_grad()
      scheduler.step()

    loss_val = loss.item()
    sum_epoch_loss += loss_val
    sum_acc += acc
    sum_f1 += f1

    if i % log_aggr == 0 or i == train_size - 1:
      print(f'[TRAIN] epoch {epoch + 1}/{num_epochs}, \
        observation {i}/{train_size} batch, \
        loss: {loss:>.4f} (avg {(sum_epoch_loss / (i + 1)):>.4f}), \
        avg acc: {(sum_acc / (i + 1)):>.4f}, \
        avg f1: {(sum_f1 / (i + 1)):>.4f}, \
        ({(len(x_batch) / (time.time() - start)):>.2f} im/s)')
    start = time.time()
```

Hàm **`train_on_epoch`** được sử dụng để huấn luyện mô hình trên mỗi epoch. Dưới đây là phân tích chi tiết của hàm này:

1. **`model.train()`**: Chuyển mô hình sang chế độ huấn luyện, điều này đảm bảo rằng các tầng trong mô hình sẽ được huấn luyện và tính toán gradient.

2. **`sum_epoch_loss = 0, sum_acc = 0, sum_f1 = 0`**: Khởi tạo các biến để tính tổng mất mát (**`sum_epoch_loss`**), tổng độ chính xác (**`sum_acc`**) và tổng F1-score (**`sum_f1`**) của từng epoch.

3. **`start = time.time()`**: Ghi nhận thời điểm bắt đầu của epoch.

4. **`train_size = len(train_loader)`**: Số lượng batch trong tập huấn luyện.

5. **`for i, (x_batch, y_batch) in enumerate(train_loader)`**: Lặp qua từng batch dữ liệu trong **`train_loader`**, trong đó **`x_batch`** là dữ liệu đầu vào và **`y_batch`** là nhãn tương ứng.

6. **`x_batch = x_batch.to(device)`**: Chuyển đổi **`x_batch`** sang thiết bị (**`device`**) được sử dụng để huấn luyện mô hình (GPU hoặc CPU).

7. **`y_batch = y_batch.to(device)`**: Chuyển đổi **`y_batch`** sang thiết bị (**`device`**) được sử dụng để huấn luyện mô hình (GPU hoặc CPU).

8. **`y_pred = model.predict('new_task', x_batch)`**: Sử dụng mô hình để dự đoán nhãn (**`y_pred`**) từ dữ liệu đầu vào (**`x_batch`**) cho tác vụ **`new_task`**.

9. **`logits = torch.exp(y_pred)`**: Áp dụng hàm exponential cho **`y_pred`** để tính toán xác suất dự đoán.

10. **`acc, f1 = evaluate(logits, y_batch)`**: Đánh giá độ chính xác (**`acc`**) và F1-score (**`f1`**) bằng cách sử dụng hàm **`evaluate`** với **`logits`** và **`y_batch`**.

11. **`loss = criteria(y_pred, y_batch)`**: Tính toán giá trị mất mát (loss) giữa **`y_pred`** và nhãn thực tế (**`y_batch`**) bằng cách sử dụng hàm **`criteria`**.

12. **`loss.backward()`**: Lan truyền ngược (backpropagation) để tính gradient của các tham số mô hình.

13. **`if i % ACCUMULATION_STEPS == 0 or i == train_size - 1`**: Kiểm tra nếu đạt đến số lượng bước tích lũy gradient (**`ACCUMULATION_STEPS`**) hoặc đạt đến batch cuối cùng trong epoch, thực hiện các bước cập nhật trọng số mô hình.

14. **`optimizer.step()`**: Cập nhật các trọng số mô hình bằng cách gọi **`optimizer.step()`**.

15. **`optimizer.zero_grad()`**: Đặt gradient về 0 bằng cách gọi **`optimizer.zero_grad()`** để chuẩn bị cho lần lan truyền tiếp theo.

16. **`scheduler.step()`**: Cập nhật learning rate bằng cách gọi **`scheduler.step()`**.

17. **`loss_val = loss.item()`**: Lấy giá trị mất mát (loss) từ tensor sang giá trị số (scalar) bằng cách sử dụng **`loss.item()`**.

18. **`sum_epoch_loss += loss_val`**: Cộng giá trị mất mát (loss) vào tổng mất mát của epoch.

19. **`sum_acc += acc`**: Cộng độ chính xác (**`acc`**) vào tổng độ chính xác của epoch.

20. **`sum_f1 += f1`**: Cộng F1-score (**`f1`**) vào tổng F1-score của epoch.

21. **`if i % log_aggr == 0 or i == train_size - 1`**: Kiểm tra nếu đạt đến số lượng batch được ghi nhận (**`log_aggr`**) hoặc đạt đến batch cuối cùng trong epoch, in thông tin về quá trình huấn luyện.

22. Cuối cùng, hàm trả về tổng mất mát trung bình (**`sum_epoch_loss / train_size`**), độ chính xác trung bình (**`sum_acc / train_size`**) và F1-score trung bình (**`sum_f1 / train_size`**) của epoch.

### Creating optimizer and learning rate schedulers

```python
param_optimizer = list(phoBERT.named_parameters())
```

Dòng mã **`param_optimizer = list(phoBERT.named_parameters())`** tạo ra một danh sách **`param_optimizer`** chứa các tên và giá trị của các tham số trong mô hình **`phoBERT`**. 

Một tham số trong mô hình được đại diện bởi một cặp (tên, giá trị). **`named_parameters()`** là một phương thức của mô hình **`phoBERT`** và trả về một trình tự chứa các cặp (tên, giá trị) của các tham số trong mô hình. 

Bằng cách sử dụng **`list()`** để chuyển đổi trình tự này thành danh sách, ta có thể truy cập và xử lý từng cặp (tên, giá trị) của các tham số trong mô hình.

```python
no_decay = ['bias', 'LayerNorm.bias', 'LayerNorm.weight']

optimizer_grouped_parameters = [
  {'params': [p for n, p in param_optimizer if not any(nd in n for nd in no_decay)], 'weight_decay': 0.01},
  {'params': [p for n, p in param_optimizer if any(nd in n for nd in no_decay)], 'weight_decay': 0.0}
]
```

Trong đoạn mã trên, chúng ta định nghĩa danh sách **`no_decay`** chứa các chuỗi tên của các tham số mà không áp dụng trọng lượng suy giảm (weight decay). Điều này bao gồm các chuỗi **"bias"**, **"LayerNorm.bias"** và **"LayerNorm.weight"**.

Sau đó, chúng ta tạo danh sách **`optimizer_grouped_parameters`** chứa hai từ điển, mỗi từ điển đại diện cho một nhóm tham số cho tối ưu hóa. 

Trong từ điển đầu tiên, chúng ta sử dụng một biểu thức danh sách (**`[p for n, p in param_optimizer if not any(nd in n for nd in no_decay)]`**) để lấy danh sách các tham số trong **`param_optimizer`** mà không có chuỗi tên nào trong **`no_decay`** xuất hiện. Các tham số này sẽ áp dụng trọng lượng suy giảm với giá trị là **0.01**.

Trong từ điển thứ hai, chúng ta sử dụng biểu thức danh sách (**`[p for n, p in param_optimizer if any(nd in n for nd in no_decay)]`**) để lấy danh sách các tham số trong **`param_optimizer`** mà có ít nhất một chuỗi tên trong **`no_decay`** xuất hiện. Các tham số này sẽ không áp dụng trọng lượng suy giảm (weight decay) với giá trị là **0.0**.

Từ điển đầu tiên đại diện cho các tham số mà chúng ta muốn áp dụng trọng lượng suy giảm, trong khi từ điển thứ hai đại diện cho các tham số mà chúng ta không muốn áp dụng trọng lượng suy giảm.

```python
num_train_optimization_steps = int(EPOCHS*len(X_train)/BATCH_SIZE/ACCUMULATION_STEPS)
```

Dòng mã **`num_train_optimization_steps = int(EPOCHS*len(X_train)/BATCH_SIZE/ACCUMULATION_STEPS)`** tính toán số bước tối ưu hóa trong quá trình huấn luyện. 

Số bước tối ưu hóa được tính bằng cách nhân số epoch (**`EPOCHS`**) với số lượng mẫu huấn luyện (**`len(X_train)`**) và chia cho kích thước batch (**`BATCH_SIZE`**) nhân với số bước tích lũy (**`ACCUMULATION_STEPS`**).

Số bước tối ưu hóa này cho biết tổng số lần cập nhật tham số mô hình sẽ được thực hiện trong quá trình huấn luyện. Bằng cách tính toán số bước này, chúng ta có thể điều chỉnh tốc độ học và lịch trình tối ưu hóa cho mô hình.

```python
optimizer = torch.optim.AdamW(optimizer_grouped_parameters, lr=LR)
scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=100, num_training_steps=num_train_optimization_steps)
scheduler0 = get_constant_schedule(optimizer)
criteria = nn.NLLLoss()
```

Trong đoạn mã trên, chúng ta định nghĩa các thành phần chính để huấn luyện mô hình:

1. **`optimizer`**: Chúng ta tạo một trình tối ưu hóa AdamW (**`torch.optim.AdamW`**) với các nhóm tham số đã được nhóm lại trong **`optimizer_grouped_parameters`**. Tham số **`lr`** được đặt bằng **`LR`**, tức là tốc độ học ban đầu.

2. **`scheduler`**: Chúng ta sử dụng hàm **`get_linear_schedule_with_warmup`** để tạo ra một lịch trình tối ưu hóa dựa trên trình tối ưu hóa và các tham số khởi tạo. Tham số **`num_warmup_steps`** được đặt là 100, tức là số bước khởi đầu trong quá trình tối ưu hóa để tăng dần tốc độ học. Tham số **`num_training_steps`** được đặt là **`num_train_optimization_steps`**, tức là tổng số bước tối ưu hóa trong quá trình huấn luyện.

3. **`scheduler0`**: Chúng ta cũng tạo một lịch trình tối ưu hóa bằng cách sử dụng hàm **`get_constant_schedule`** để duy trì tốc độ học không đổi cho tất cả các bước tối ưu hóa.

4. **`criteria`**: Chúng ta tạo một đối tượng **`criteria`** từ lớp **`NLLLoss`** trong module **`torch.nn`**. Đối tượng này sẽ được sử dụng để tính toán hàm mất mát (loss) trong quá trình huấn luyện.

```python
splits = list(StratifiedShuffleSplit(n_splits=1, random_state=123, test_size=0.1).split(X_train, y))
```

Sử dụng lớp **`StratifiedShuffleSplit`** từ thư viện scikit-learn để tạo các phân chia huấn luyện và kiểm tra sử dụng phương pháp lấy mẫu có độ đồng đều. Đây là cách hoạt động của đoạn mã:

- **`StratifiedShuffleSplit(n_splits=1, random_state=123, test_size=0.1)`** tạo một đối tượng của lớp **`StratifiedShuffleSplit`**.
  - **`n_splits=1`** xác định rằng chỉ tạo một phân chia huấn luyện và kiểm tra.
  - **`random_state=123`** đặt giá trị hạt giống ngẫu nhiên để đảm bảo có thể tái tạo kết quả phân chia.
  - **`test_size=0.1`** xác định tỷ lệ dữ liệu dùng cho kiểm tra. Ở đây, 10% dữ liệu sẽ được sử dụng cho kiểm tra.

- **`.split(X_train, y)`** gọi phương thức **`split`** trên đối tượng **`StratifiedShuffleSplit`**.
  - **`X_train`** đại diện cho dữ liệu đặc trưng muốn phân chia.
  - **`y`** đại diện cho biến mục tiêu (nhãn) liên quan đến dữ liệu đặc trưng.

- **`list(...)`** chuyển đổi trình tạo ra bởi **`split`** thành một danh sách.
  - Phương thức **`split`** trả về một bộ tạo lặp được sử dụng để tạo các chỉ số cho các phân chia huấn luyện và kiểm tra. Bằng cách chuyển đổi nó thành danh sách, bạn có thể truy cập trực tiếp các chỉ số đã được tạo ra.

```python
for fold, (train_idx, val_idx) in enumerate(splits):
  best_score = 0
  print("Training for fold {}".format(fold))

  train_dataset = torch.utils.data.TensorDataset(torch.tensor(X_train[train_idx],dtype=torch.long), torch.tensor(y[train_idx],dtype=torch.long))
  valid_dataset = torch.utils.data.TensorDataset(torch.tensor(X_train[val_idx],dtype=torch.long), torch.tensor(y[val_idx],dtype=torch.long))
  print('train size:', len(train_dataset))
  print('valid size:', len(valid_dataset))

  frozen = True

  for epoch in tqdm(range(EPOCHS + 1)):
    
    if epoch > 0 and frozen:
      for child in phoBERT.children():
        for param in child.parameters():
          param.requires_grad = True
      frozen = False
      del scheduler0
      torch.cuda.empty_cache()

    train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
    valid_loader = torch.utils.data.DataLoader(valid_dataset, batch_size=BATCH_SIZE, shuffle=False)

    print('\nEpoch: ', epoch + 1)
    optimizer.zero_grad()
    train_on_epoch(train_loader=train_loader,
                  model=phoBERT,
                  optimizer=optimizer, 
                  epoch=epoch, 
                  num_epochs=EPOCHS + 1, 
                  criteria=criteria, 
                  device=DEVICE, 
                  scheduler=(scheduler0 if frozen else scheduler))
    
    acc, f1, loss = validate(valid_loader, phoBERT, device=DEVICE)
    print('Epoch {} validation acc: {:.4f}, f1: {:.4f}, loss: {:.4f} \n'.format(epoch + 1, acc, f1, loss))
    if f1 >= best_score:
      torch.save(phoBERT.state_dict(), os.path.join(CKPT_PATH, f'model.bin'))
      best_score = f1
```

Đoạn mã trên được hiểu như sau:

1. **`enumerate(splits)`**: Được sử dụng để lặp qua các phần tử của biến **`splits`** cùng với chỉ số của chúng.

2. Bên trong vòng lặp, mã khởi tạo biến **`best_score`** để theo dõi điểm F1 tốt nhất đã đạt được cho fold hiện tại.

3. Các tập dữ liệu huấn luyện và validation được tạo ra bằng cách sử dụng các chỉ số huấn luyện và validation được lấy từ fold hiện tại.

4. Biến **`frozen`** được đặt thành **`True`**, cho biết các tham số của mô hình ban đầu đang bị đóng băng (không huấn luyện được).

5. Mã tiếp tục bước lặp qua từng epoch, bao gồm một epoch bổ sung (**`EPOCHS + 1`**). Mục đích của epoch bổ sung này là để mở đóng băng các tham số của mô hình sau các epoch ban đầu.

6. Bên trong vòng lặp epoch, nếu epoch hiện tại lớn hơn 0 và **`frozen`** là **`True`**, các tham số của mô hình được mở đóng băng bằng cách đặt **`requires_grad`** thành **`True`** cho tất cả các tham số.

7. Các data loader huấn luyện và validation được tạo ra bằng cách sử dụng các tập dữ liệu tương ứng.

8. Hàm **`train_on_epoch`** được gọi để thực hiện huấn luyện trên epoch hiện tại.

9. Sau khi huấn luyện epoch, hàm **`validate`** được gọi để đánh giá hiệu suất của mô hình trên tập validation.

10. Các độ đo validation, bao gồm độ chính xác (**`acc`**), F1 score (**`f1`**), và loss (**`loss`**), được in ra màn hình.

11. Nếu F1 score hiện tại (**`f1`**) tốt hơn điểm số tốt nhất trước đó (**`best_score`**), bảng điều khiển trạng thái của mô hình được lưu vào một tệp tin.

2. Tạo một **`test_dataset`** từ dữ liệu kiểm tra (**`X_test`**) và đóng gói nó vào một **`test_loader`**. **`test_loader`** được sử dụng để lặp qua các mini-batch từ tập dữ liệu kiểm tra.

3. **`phoBERT.eval()`** được gọi để chuyển sang chế độ đánh giá (evaluation mode) của mô hình. Điều này thông báo cho mô hình rằng nó sẽ được sử dụng để dự đoán và không được huấn luyện thêm.

4. Trong vòng lặp **`for i, (x_batch,) in tqdm(enumerate(test_loader), total=len(test_loader))`**, mỗi mini-batch **`x_batch`** từ **`test_loader`** được lặp qua.

5. **`logits = phoBERT.predict('new_task', x_batch)`** được sử dụng để dự đoán đầu ra logits cho mini-batch hiện tại bằng cách gọi phương thức **`predict`** trên mô hình **`phoBERT`**. Các logits là các giá trị đầu ra trước khi được chuyển qua hàm softmax.

6. **`probs = torch.exp(logits)`** được sử dụng để tính toán xác suất dự đoán bằng cách áp dụng hàm softmax trên logits.

7. **`y_pred = np.argmax(probs.detach().cpu().numpy(), axis=1)`** được sử dụng để chọn lớp có xác suất cao nhất từ các xác suất dự đoán. **`detach()`** được sử dụng để tách các giá trị gradient, và **`cpu().numpy()`** được sử dụng để chuyển đổi kết quả thành mảng NumPy.

8. **`preds = np.concatenate([preds, y_pred])`** được sử dụng để ghép các dự đoán của mini-batch hiện tại vào mảng **`preds`**. Mảng **`preds`** chứa tất cả các dự đoán trên toàn bộ tập dữ liệu kiểm tra.
